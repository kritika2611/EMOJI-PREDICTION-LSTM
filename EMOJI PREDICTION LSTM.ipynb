## DATASET

```python
import pandas as pd
```

```python
import numpy as np
```

```python
train=pd.read_csv('EmojifiedText/emojify_train_x.csv',header=None)
label=pd.read_csv('EmojifiedText/Emojify_Y_train.csv',header=None)
```

```python
train.head()
```

```python
train=train.values
```

```python
train
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tok=RegexpTokenizer('[a-zA-Z ]+')
```

```python
X_train=[]
for i in range(train.shape[0]):
    lis=tok.tokenize(train[i][0])
    for sent in lis:
        if sent!=' ':
            X_train.append(sent)
        
```

```python
X_train
```

```python
label.values
```

```python
Y_train=[]
tok2=RegexpTokenizer('[0-9]+')
for i in range(label.shape[0]):
    lis=tok2.tokenize(label[0][i])
    for v in lis:
        Y_train.append(v)
```

```python
print(Y_train)
```

## CONVERTING SENTENCES INTO EMBEDDINGS

```python
f=open('glove.6B.50d.txt',encoding='utf-8')
```

```python
embeddings={}
for l in f:
    lis=l.split(' ')
    word=lis[0]
    coef=np.asarray(lis[1:],dtype=float)
    embeddings[word]=coef
f.close()
```

```python
len(embeddings.keys())
```

```python
embeddings['to']
```

## CONVERTING WORDS INTO VEC(embedding layer output)

```python
def emb_out(X):
    out=np.zeros((len(X),10,50))
    for i,s in enumerate(X):
        words=s.split(' ')
        for j,w in enumerate(words):
            try:
                out[i][j]=embeddings[w.lower()]
            except:
                out[i][j]=np.zeros((50,))
            
    return out
```

```python
embeddings_X_train=emb_out(X_train)
```

```python
embeddings_X_train.shape
```

## LSTM MODEL

**define model architecture**

```python
#converting Y_train into one hot
from keras.utils import to_categorical
Y_train_cat=to_categorical(Y_train)
```

```python
Y_train_cat
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Dropout,Dense
from keras.callbacks import ModelCheckpoint
```

```python
check_point=ModelCheckpoint('EMOJI_MODEL/best_model_2.h5', monitor='val_loss', verbose=True, save_best_only=True)
```

```python
model=Sequential()
model.add(LSTM(64,input_shape=(10,50)))
model.add(Dropout(0.4))
model.add(Dense(5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc'])
```

```python
hist=model.fit(embeddings_X_train,Y_train_cat,epochs=100,batch_size=32,shuffle=True,callbacks=[check_point],validation_split=0.2)
```

```python
data=hist.history
```

```python
import matplotlib.pyplot as plt
plt.plot(data['loss'],label='Training_loss')
plt.plot(data['val_loss'],label='Validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
plt.plot(data['acc'],label='Training accuracy')
plt.plot(data['val_acc'],label='Validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
model.save('EMOJI_MODEL/my_model2.h5')
```

# STACKED LSTM

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
```

```python
m=Sequential()
#defining model
m.add(LSTM(64,input_shape=(10,50),return_sequences=True))
m.add(Dropout(0.4))
m.add(LSTM(64,return_sequences=False))
m.add(Dropout(0.4))
m.add(Dense(5,activation='softmax'))
```

```python
m.summary()
```

```python
m.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['acc'])
```
